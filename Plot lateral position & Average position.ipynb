```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
l_c = open('/home/m/Dropbox/HMM/l_c_Train.csv','r')
D = [s.split(',') for s in l_c.readlines()]
Data=[[]]
count=0
for i in D:
    if i[0]=='Lateral position':
        Data.append([])
        count=count+1
        continue

    Data[count].append(i)
#print(Data)
N = len(Data)
#print(N)
for j in range(N):
    Data[j] = np.array([[float(k) for k in data] for data in Data[j]])

T = np.min([d.shape[0] for d in Data])

for j in range(N):
    Data[j] = Data[j][-T:]
   # print (Data[j].shape)
## pyplot.axis([X min ,X max ,Y min ,Y max ])

fig = plt.figure()
plt.axis([0, 150,11, 0])
plt.xlabel('time')
plt.ylabel('Lateral position')
plt.title('variation of lateral position in Lane Changing') 
#plt.legend('direction) 
ave=[]
for i in range(N):
    s=[]
    for j in range(150):
        s.append(Data[i][j][0])
    plt.plot(range(len(s)), s)
    ave.append(s)
a=[]
sum=0
for i in range(len(ave[1])):
    for j in range(25):
        sum = sum+ave[j][i]
    av=sum/25
    a.append(av)
    sum=0
#print(a)
plt.plot(range(len(s)), a,linewidth=7.3,linestyle='dashed')
```
